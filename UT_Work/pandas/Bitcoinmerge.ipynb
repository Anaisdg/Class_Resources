```python
import pandas as pd 
file_path_one = "bitcoin_cash_price.csv"
file_path_two= "dash_price.csv"

bitcoin = pd.read_csv(file_path_one)
dash = pd.read_csv(file_path_two)
```

```python
bitcoin.head()
```

```python
dash.head()
```

```python
bitcoin = bitcoin.rename(columns={'Open': 'Bitcoin Open','High':'Bitcoin High', 'Low':'Bitcoin Low', 'Close':'Bitcoin Close', 'Volume': 'Bitcoin Volume', 'Market Cap':'Bitcoin Market Cap'})
bitcoin.head()
         
```

```python
dash = dash.rename(columns={'Open': 'Dash Open','High':'Dash High', 'Low':'Dash Low', 'Close':'Dash Close', 'Volume': 'Dash Volume', 'Market Cap':'Dash Market Cap'})
dash.head()
```

```python
combine = pd.merge(bitcoin,dash, on='Date')
combine.head()
```

```python
bitcoin_open = combine['Bitcoin Open'].max()
dash_open = combine['Dash Open'].max()
bitcoin_totv = round(combine['Bitcoin Volume'].sum()/1000000000, 2)
dash_totv = round(combine['Dash Volume'].sum()/1000000000, 2)
```
